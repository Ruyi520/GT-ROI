# Target Labels

From communication with Robert Bixler (02/16/2019), the `FirstReportType` and `NumberOfReports` feature columns
basically hold the target/label to be used for the ML model.  The `NumberOfReports`, I believe, gives the basic
label we would use intially.  So when this is 1 or `self-caught` the subject reported mind wandering during
the given trial, and when 0, they did not.

```python
t = d['FirstReportType']
```

```python
t.astype('category').describe()
```

```python
pd.factorize(d['FirstReportType'])
```

```python
pd.factorize(d['NumberOfReports'])
```

```python
d['NumberOfReports'].describe()
```

```python
d['FirstReportType'].describe()
```

If this is our label, just over 25% of the trials are instances of reported mind windering, and the other 75%
did not involve a reported instance of mind wandering.  This might be problematic for some of the reported
measures in the paper, because this means that a baseline classifier that always reports no mind wandering should
achieve 75% success on this data.  Thus classifiers should beat a baseline performance of 75% or else they
are not really performing anything.

```python
np.sum(d['FirstReportType'] == 'none')
```

```python
np.sum(d['FirstReportType'] == 'self-caught')
```

If what I was told is true, then `FirstReportType` and `NumberOfReports` should be equivalent.  Whenever
`FirstReportType` is 0, then `NumberOfReports` should be 0, and when `FirstReportType` is non-zero, then
`NumberOfReports` should be 1.

```python
first_report = pd.factorize(d['FirstReportType'])[0]

print(np.sum(first_report == 0))
print(np.sum(first_report == 1))
print(np.sum(first_report == 2))

number_reports = pd.factorize(d['NumberOfReports'])[0]

# test that everyplace FirstReportType is 0 then NumberOfReports is 0
print(np.alltrue( (first_report == 0) == (number_reports == 0) ))
```
