# Select and Create Clean Images for Training

Images in `tf_data` directory are a set of 43937 images of peoples faces/heads in different orientations.

Image file names are in format `NAME_f2_x1_x2.jpg`

The first field in the image names appears to be in 1 of 3 sets `['AFW', 'HELEN' 'IBUG']`

Actually the IBUG patterns appear to have some that don't follow the naming pattern convention.
 bn
 
 ]
The most general pattern is `IBUG_image_xxx_y_z`  where xxx is a 3 digit number filled with 0 if number
is less than 100, e.g. 061 or 129.  Sometimes this field is not 0 filled, so there are some `IBUG_image_09`
names which may have been image 009 mislabeled.  Likewise there are some `IBUG_image_99` names.  Perhaps
both of these should be image 099?  This appears to be the only set of images with `IBUG` names that
dont use 3 digits in the field 3.  Also for the `IBUG` images some have 2 numbers `_y_z` at end where
first one `y` is usually a single digit 1 or 2, and second sometimes ranges to 2 digits like up to 15 or 20.
But some of the `IBUG` images only have a single numeric field after the 3 digit id.

Looking at the `AFW` images, I believe they always have `_y_z` where `y` is always 1 or 2.  Images with the same
`NAME` and field 2 id with a 1 in field 3 are the same image rotated.  If field 3 is a 2 it appears to be
a different person rotated.

So I think for the `AFW` and `HELEN` images, the field 3 is also part of the id identifying a different person
being rotated.  For example, `HELEN_288343911` has a set of `_1_0` to `_1_14` which are rotations of one of 2
people in the image.  then this same prefix with numbers `_2_0` to `_2_14` are rotations of the other person
in the image.  And for example `HELEN_288344083` has 3 people in the image, and thus the third field has numbers
1, 2 and 3.

In general I think that for the `AFW` and `HELEN` images, the first 3 fields are a unique id of a person
(from a group of possible 2 or 3 people), and the 4th field is a sequence of rotations of that persons face.
For the `IBUG_image` files the 3rd appears to be mostly a number from 000 to 119.  If there is only 1 other field
after this it is most likely a rotation number.  If there are 2 number fields after this one then again the
first one is like a different person.


So in short, the pattern of filenames is different.  For `AFW` and `HELEN` files I think all of them
fit this pattern

`([AFW|HELEN]_[0-9]+_[1-9])_([0-9]+).jpg`

The first 3 fields of name, some id number and number 1,2,3 identifies the person (or 1 of some group of 2 or 3
people) in the photo to be manipulated.  And the last field of 1 or 2 digits is a rotation manipulation sequence
number.

The `IBUG_image` files are not quite as consistent in the names.  First of all, I consider `IBUG_image` as the
name of this category.  So one pattern looks like this

`([IBUG_image]_[0-9]+_[0-9]+)_([0-9]+).jpg`

This is similar to before, though the first number field is usually an id of 3 digits, but after the name and
3 digit number is anohter number indicating a person number, and the last field is a rotation manipulation.

But some of the `IBUG_image` do not have the person number so they have a pattern of


`([IBUG_image]_[0-9]+)_([0-9]+).jpg`

e.g. they are simply the name and a 3 digit id number, with one last field being the rotation manipulation.

I believe the last field is always the rotation number.  So a simplified patter that I think matches all is
that all characters up to the last _ are the first field which is an id of the person being rotated, then
the last field after the last _ is the rotation number.

`([A-Z0-9_]+)_([0-9]+).jpg`

```python
tf_dir = '../data/tf_data/'

# some data exploration, confirm number of images and that we only have 3
# categories of items in the filenames
categories = set()
category_re = r'([a-zA-Z_]+)_([0-9_]+)\.jpg'

num_images = 0
num_excluded = 0

for image_name in os.listdir(tf_dir):
    # extract the initial alphabetic field from the end numeric fields
    # using regular expression
    match = re.match(category_re, image_name)
    
    # determine unique categories we observe in filenames
    if match:
        categories.add(match.group(1))
        num_images += 1
    else:
        print("Excluding file: ", image_name)
        num_excluded += 1
        
print("Number of images processed: ", num_images)
print("Number of images excluded : ", num_excluded)
print("Unique category/names in field 1 of image names: ", categories)
```

## Select N Random Images for Training

In original notebook first set of code for creating clean images for training was selecting 50 images
and copying them to the `clean_data` directory.  The regular expression being used caused the
`IBUG_image` images to not be considered for selection, which I am not sure was intentional or not.
I modified a bit to select N images at random, but considering only the `AWS` and `HELEN` set as was
originally being done.  Also not sure, but it might be worthwhile using only the initial rotation 0
images?  These faces still have varying pitch, roll and yaw, and have not been digitally rotated, so might
be cleaner?  Alternatively, really what we want is to learn how the rotations were done here for this
image set.  Because if we want to do date generation correctly for this problem, we can't really use the
Keras image data generators, which are simply doing things like rotating and shifting images.  We maybe need to
do that, but we also want to digitally/artificailly change the pitch, roll and yaw of a source image to
generate data with a wider variety of these training parameters for input.

We also have a constraint that we would like only 1 face in each image.  So to ensure we still have exactly
N images to train with, we also add in check here to detect faces, and only select images that
meet this constraint.

```python
# parameterize all of this work into a function so that we can reuse it to
# create separate train, validate and test data sets
def select_images_for_dataset(src_dir, dst_dir, num_images_to_select):
    """Select a number of images at random for use as a data set and copy the images over
    to subdirectories so we can clean them and set them up for a Keras data generator to flow
    from directories.
    
    Parameters
    ----------
    src_dir - The source directory of images to randomly select from.
    dst_dir - The destination directory to copy selected images to
    num_images_to_select - The number of images to select at random for the dataset being
      created.
      
    """
    # DANGER: we remove any existing clean_data directory here, so don't run this cell unless you really want
    # to select another set of N images for training and testing
    if os.path.isdir(dst_dir):
        print("Removing old training set that was detected: ", dst_dir)
        shutil.rmtree(dst_dir)
    os.makedirs(dst_dir)


    # Make a list of image file names we will consider for random selection
    candidate_images_re = r'([AWS|HELEN]).*\.jpg' # should this be a parameter to the function?
    candidate_images = []
    for image_name in os.listdir(src_dir):
        # determine if image is candidate for random selection
        match = re.match(candidate_images_re, image_name)
        if match:
            candidate_images.append(image_name)

    print("Number of image candidates for random selection: ", len(candidate_images))

    # randomly shuffle images and select top N images to be copied
    random.shuffle(candidate_images)

    # use amazon rekognition service to detect number of faces in image
    session = boto3.Session(profile_name='default')
    client = session.client('rekognition')

    num_images_selected = 0
    image_no = 0
    while num_images_selected < num_images_to_select:
        image_name = candidate_images[image_no]
        image_no += 1

        src_file_path = os.path.join(src_dir, image_name)

        # check for a single face in each image
        image = open(src_file_path, 'rb')
        response = client.detect_faces(Image={'Bytes': image.read()})
        label = response['FaceDetails']
        numFaces = len(label)

        if numFaces == 1:
            num_images_selected += 1
            dst_file_name = "face-%03d.jpg" % (num_images_selected)
            dst_file_path = os.path.join(dst_dir, dst_file_name)

            print("Copy %s to %s" % (src_file_path, dst_file_path))
            shutil.copyfile(src_file_path, dst_file_path)

            # save information about original image in case needed
            image_label_filename = 'face-%03d-label.txt' % (num_images_selected)
            image_label_path = os.path.join(dst_dir, image_label_filename)
            label_file = open(image_label_path, 'a')
            label_file.write('original: %s\n' % (image_name))
            label_file.close()

            

# randomly select test, train and validation data sets
src_dir = '../data/tf_data'

print("Selecting train data set")
train_dir = '../data/head_pose_data/train'
num_train_images = 100
select_images_for_dataset(src_dir, train_dir, num_train_images)
print("\n\n")

print("Selecting validation data set")
validation_dir = '../data/head_pose_data/validation'
num_validation_images = 50
select_images_for_dataset(src_dir, validation_dir, num_validation_images)
print("\n\n")

print("Selecting test data set")
test_dir = '../data/head_pose_data/test'
num_test_images = 50
select_images_for_dataset(src_dir, test_dir, num_test_images)
print("\n\n")

```

# Determine Image Labels

Using AWS `rekognition` image processing service, determine training labels of our training images.

```python
# again parameterize this into a function so we can reuse to determine labels for
# test/train/validation sets selected and put into different directories
def determine_dataset_labels(dataset_dir):
    """Given a directory of selected head pose images, determine the pose labels.  This
    function determines pose labels, extracts a new image using a tight bounding box only around
    the detected face in the original image.  And the pitch, roll and yaw labels are written to
    a corresponding file so they can be used by a Keras data flow generator to supply
    labels for the images.
    
    Parameters
    ----------
    dataset_dir - A file path, the location of the dataset images.
    
    Returns
    -------
    label_dict - Returns a dictionary with the image number as the key, and 3 pieces of information
       for each image, the pose labels, the filname of the cropped image to be used and the actual cropped
       image as a numpy array.
    """
    # DANGER: we remove any existing directory here, so don't run this cell unless you really want
    # to select another set of N images for training and testing
    # we use 'faces' subdirectory to put the final cropped images and pose labels we generate here
    dst_dir = os.path.join(dataset_dir, 'faces')
    if os.path.isdir(dst_dir):
        print("Removing old data set that was detected: ", dst_dir)
        shutil.rmtree(dst_dir)
    os.makedirs(dst_dir)
    
    # use amazon rekognition service to determine pose labels
    session = boto3.Session(profile_name='default')
    client = session.client('rekognition')

    # all images seem to be 450x450 pixels, though maybe we should detect this for each image instead
    # of hardcoding assumption of image size
    width, height = 450, 450

    # regular expression, only use the the original face images, before cropping, etc.
    input_image_re = r'face-(\d+)\.jpg'

    # a dictionary accessed using the image number as key to hold resulting
    # labels and other information
    label_dict = {}

    for image_name in os.listdir(dataset_dir):
        match = re.match(input_image_re, image_name)
        if match:
            old_photo = os.path.join(dataset_dir, image_name)
            image = open(old_photo, 'rb')

            # determine image labels using rekognition engine
            response = client.detect_faces(Image={'Bytes': image.read()})
            label = response['FaceDetails'][0] # should only be 1 face, so get index 0 info from response

            # extract pitch, roll, yaw labels
            pose_labels = (int(label['Pose']['Roll']),
                           int(label['Pose']['Yaw']),
                           int(label['Pose']['Pitch'])
                          )


            # extract bounding box of face to crop
            x, y, w, h = (int(label['BoundingBox']['Left'] * width),
                          int(label['BoundingBox']['Top'] * height),
                          int(label['BoundingBox']['Width'] * width),
                          int(label['BoundingBox']['Height'] * height),
                         )

            # convert bounding box format and crop to face of each image
            x1, y1, x2, y2 = x, y, (x+w), (y+h)
            cropped_photo = cv2.imread(old_photo)
            new_photo = cropped_photo[y1:y2, x1:x2]

            # save cropped version
            image_id = int(match.group(1))
            new_photo_filename = 'face-%03d-cropped.jpg' % (image_id)
            new_photo_path = os.path.join(dst_dir, new_photo_filename)
            cv2.imwrite(new_photo_path, new_photo)
            
            # save pose label data to separate label file
            image_label_filename = 'face-%03d-label.txt' % (image_id)
            image_label_path = os.path.join(dataset_dir, image_label_filename)
            label_file = open(image_label_path, 'a')
            label_file.write('roll: %d\n' % (pose_labels[0]))
            label_file.write('yaw: %d\n' % (pose_labels[1]))
            label_file.write('pitch: %d\n' % (pose_labels[2]))
            label_file.write('crop_width: %d\n' % (w))
            label_file.write('crop_height: %d\n' % (h))
            label_file.close()

            # insert into dictionary information we need about labeled training set
            label_dict[image_id] = (pose_labels, new_photo_filename, new_photo)

            print("%20s: roll=%d yaw=%d pitch=%d" % 
                  (new_photo_filename, pose_labels[0], pose_labels[1], pose_labels[2]))
            
    return label_dict
    
    
# determine labels for the train dataset
print("Determine pose labels for the training dataset...")
train_dir = '../data/head_pose_data/train'
train_labels = determine_dataset_labels(train_dir)
print("\n\n")

# determine labels for the validation dataset
print("Determine pose labels for the validation dataset...")
validation_dir = '../data/head_pose_data/validation'
validation_labels = determine_dataset_labels(validation_dir)
print("\n\n")

# determine labels for the test dataset
print("Determine pose labels for the testing dataset...")
test_dir = '../data/head_pose_data/test'
test_labels = determine_dataset_labels(test_dir)
print("\n\n")

```

# Deep Learning Model Creation and Training

Create a model using a pretrained VGG base.  The goal is to learn the pitch, roll and yaw labels for cropped faces.

First of all lets make sure we can create an appropriate flow data generator from our head pose image
datasets.  The labels are not binary or categorical here.  We need to get the 3 labels for each image
that we flow from directory.

```python
# rescale here simply converts from uint8 0-255 values for RGB pixel colors to a real value from 0.0 to 1.0
datagen = ImageDataGenerator(rescale=1.0/255.0)
batch_size = 10

def extract_features(directory, sample_count):
    """Extract features and labels (and return them) from a directory of input
    images.
    
    Parameters
    ----------
    directory - The directory where the inputs are from.
    sample_count - The number of images to process
    
    Returns
    -------
    features - The features for input into a network (e.g. the image as a numpy array)
    labels - The outputs, here the roll, yaw, pitch label for each input image
    """
    width, height = 100, 100 # target width,height of images to rescale to for input
    features = np.zeros(shape=(sample_count, width, height, 3))
    labels = np.zeros(shape=(sample_count, 3))
    
    # create a generator to flow images from directory and yield data as numpy array of correct
    # shape
    generator = datagen.flow_from_directory(
        directory,
        target_size=(width, height),
        batch_size=batch_size,
        class_mode=None,
        shuffle=False # important, causes image_id 1 to be first to flow, followed by 2,3,...
    )
    
    # use the generator to extract the image data into our features array for input training
    i = 0
    for inputs_batch in generator:
        features[i * batch_size : (i+1) * batch_size] = inputs_batch
        i += 1
        if i * batch_size >= sample_count:
            break
    
    # now process the label files ourselves to extract the pitch,roll and yaw pose labels we need
    for image_id in range(1, sample_count + 1):
        image_file_name = "face-%03d-label.txt" % (image_id)
        image_file_path = os.path.join(directory, image_file_name)
        
        pitch = 0
        roll = 0
        yaw = 0
        for line in fileinput.input(image_file_path):
            key, value = line.split(':')
            if key == 'roll':
                roll = int(value)
            if key == 'yaw':
                yaw = int(value)
            if key == 'pitch':
                pitch = int(value)
        labels[image_id-1] = np.array([roll, yaw, pitch])
        
    # return the extracted input features and output labels
    return features, labels

base_dir = '../data/head_pose_data'

train_dir = os.path.join(base_dir, 'train')
num_train_images = 100
train_features, train_labels = extract_features(train_dir, num_train_images)

validation_dir = os.path.join(base_dir, 'validation')
num_validation_images = 50
validation_features, validation_labels = extract_features(validation_dir, num_validation_images)

test_dir = os.path.join(base_dir, 'test')
num_test_images = 50
test_features, test_labels = extract_features(test_dir, num_test_images)
```

```python
# Two-stage convolution network model
# load VGG base, set as not trainable, we don't include the top so we can add on our own trainable layers
# for the pose labeling task
vgg_base = keras.applications.VGG16(input_shape=(100,100,3), include_top=False, weights='imagenet')
vgg_base.trainable = False
vgg_base.summary()

# create layers that will be trained on the head pose data
model = Sequential()
model.add(Flatten())
model.add(Dense(1024, activation='relu', input_dim=3 * 3 * 512))
model.add(Dropout(0.5))
model.add(Dense(3, activation='linear'))

#model.summary()

# combine the models
head_pose_model = Sequential([vgg_base, model])

head_pose_model.summary()
```

```python
head_pose_model.compile(optimizer=optimizers.RMSprop(lr=2e-5),
                       loss='mean_squared_error',
                       metrics=['mse'])
```

```python
history = head_pose_model.fit(train_features, #train_labels, 
                             epochs=10,
                             batch_size=5,
                             validation_data=(validation_features, validation_labels))
```

```python
mse = history.history['mse']
val_mse = history.history['val_mse']

epochs = range(1, len(mse) + 1)

plt.figure(figsize=(15,10))
plt.plot(epochs, mse, 'bo', label='Training mse')
plt.plot(epochs, val_mse, 'b', label='Validation mse')
plt.title('Training and validation Mean Square Error (mse)')
plt.legend();
```

```python
# This function is a spiral function that returns an array of tuple pairs
# used for generating our mouse position (x,y) values
def spiral(X, Y):
    pairs = []
    x = y = 0
    dx = 0
    dy = -5
    for i in range(max(X, Y)**2):
        if (-X/2 < x <= X/2) and (-Y/2 < y <= Y/2):
            pairs.append((x, y))
        if x == y or (x < 0 and x == -y) or (x > 0 and x == 5-y):
            dx, dy = -dy, dx
        x, y = x+dx, y+dy
    return pairs
```

```python
# We define our eye aspect ratio below where 0=< ratio =< 1 for closed to opened respectively
def Ear(eye):
    # compute the euclidean distances between the two sets of
    # vertical eye landmarks (x, y)-coordinates
    A = LA.norm(eye[1] - eye[5])
    B = LA.norm(eye[2] - eye[4])

    # compute the euclidean distance between the horizontal
    # eye landmark (x, y)-coordinates
    C = LA.norm(eye[0] - eye[3])

    # compute the eye aspect ratio
    ratio = (A + B) / (2.0 * C)

    # return the eye aspect ratio
    return ratio
```

## Create an Classroom object that parameterizes calculations of a room seen through a camera.
Given measured parameters of calibrated points in an image, and certain assumptions
about visible heads/faces in a camera image, encapsulate calculations that transform measurements
from images into an external cartesian coordinate system.  Coordinates originate at top corner of classroom
(x,y,z = 0,0,0)

```python
import cv2
import numpy as np
```

```python
class Classroom:
    """Classroom object to encapsulate calculations of coordinate system of objects seen in an image.
    The Classroom object is calibrated to pixels in the image, and also uses assumptions about
    the size of faces/heads identified in the image to calculate
    """
    
    def __init__(self, width, depth, height, offset):
        """Constructor for the Classroom object.  The parameters width, height and depth are the measurements
        of the classroom being viewed by the camera.  Camera is positioned at front of the classroom.
        The cartesian system y axis corresponds to left and right from camera point of view.  The width
        ofthe classroom is measurement of the y axis in the cartesian system originating at the right
        corner of the room from camera point of view.  Depth corresponds to the direction moving towards
        and away from the camera and is designated as the x axis.  Height designates the height from floor to
        ceiling of the classroom.  
        
        Parameters
        ----------
        width - Width of classroom this object is making calculations with.  Corresponds to the y axis
            dimension and visually represents directions left and right from the view of the camera.
        height - Height of classroom this object encapsulates.  The average height from ceiling to floor of
            the classroom.  This corresponds to up and down in the camera point of view and is designated the
            z axis in the cartesian coordinate system.
        depth - Depth of classroom in the direction towards and away from the camera.  This is designated
            as the x axis in the cartesian coordinate system.  Objects move closer or further from the
            camera in this dimension from the camera point of view.
        offset - The depth distance between image plane and the ground plane
        """
        # keep track of measured calibration parameters of the classroom
        self.width = width
        self.height = height
        self.depth = depth
        self.offset = offset
        
        # x, y, z are the cartesian coordinate location of the camera in the room
        # perhaps x,y,z need to be parameters as well, or else this object forces camera to be
        # positioned exactly in center of room.
        self.x = 0 # is this a bug in original code, seems like if x,y,z is camera position, camera is at 0 on x axis
        self.y = self.width / 2.0  # this assume camera is exactly centered on front wall of classroom in middle of y direction
        self.z = self.height / 2.0 # again this would assume camera position is half way between floor and ceiling
        
    def get_floor_homography(self, detection_plane):
        """In this function we calculate a homography for the plane represented by the floor of the classroom.
        A homography maps points in 1 image to points in another.  Here we assume one image is from straight
        above looking down at the floor, which is the cartesian plane for the x and y dimensions of the
        classroom we want to estimate.  The other image will be pixel locations in our image that correspond 
        to the 4 cornors of the room.
        
        Discussion, we will not be able to see the two corners in the image the camera is against.  So don't we
        really need to define a plane with all 4 points visible in the camera image on the floor?
        
        Parameters
        ----------
        detection_plane - We expect a list of 4 tupples.  These represent the 4 corners of the room identified
            in a calibration step as pixel locations from a still image from our placed camera.
            These should correspond to our defined cartesian coordinates of our classroom in the expected
            order we need.
            
        Returns
        -------
        floor_homography - This function returns the calculated homography matrix back to the caller
            and also saves it as self.floor_homography in the object as well in case needed
        """
        # a homography calculation works by taking 4 points in each of 2 images (coordinate systems)
        # that define corresponding points on a plane.  The result is a 3x3 transformation matrix
        # known as a homography.  W

        # remember the detection_plane used in case needed again
        self.detection_plane = detection_plane # detection plane points need to be ordered in same order as class plane

        # create a list of the 4 points in the cartesian coordinates we are defining for the classroom.
        p1 = [0, 0] # origin point, to left of camera view along wall where camera located
        p2 = [width, 0] # x is width of classroom, point to right of camera along wall where camera located
        p3 = [0, depth] # x is 0 y is at depth of classroom, so point at back wall of classroom, to left of camera view
        p4 = [width, depth] # The corner at back wall to right in camera view
        self.class_plane = np.array([p1, p2, p3, p4])
        
        # use cv2 to find the homography matrix for us
        floor_homography, status = cv2.findHomography(self.detection_plane, self.class_plane)
        self.floor_homography = floor_homography
        
        return floor_homography
    
    def camera_intrinsic_matrix(detection_plane, class_plane, image_size):
        """ The floor homography calculated includes the properties of the camera which 
        is defined as the camera intrinsic matrix K. 
        Paramters
        ---------
        
        Returns
        -------
        K - (3 X 3) shaped numpy of the camera instric properties
        """
        reprojection_error, camera_matrix, distortion_coefficient, rotation_v,\
        translation_v = cv2.calibrateCamera(detection_plane, class_plane, image_size)
        
        # units of pixels per millimeter in the x-axis
        sx = (class_plane[1][0] - class_plane[0][0])/(detection_plane[1][0] - detection_plane[0][0])

        # units of pixels per millimeter in the y-axis
        sy =  (class_plane[2][1] - class_plane[1][1])/(detection_plane[2][1] - detection_plane[1][1])
        
        F = 3.99 #focal length of an iphone 8 camera 
        fx, fy = F*sx, F*sy
        W, H = img.shape[:2]
        cx, cy = W//2, H//2
        print(sx, sy)
        
        self.K = np.array([[fx, 0, cx], [0, fy, cy], [0, 0, 1]])
        self.K_inv = np.linalg.inv(K)
        return K, K_inv
    
    
    def map_floor_pixel(self, src_pixels):
        """Given a pixel that is located on the floor of our classroom image, return the
        coordinates of the mapped pixel to the cartesian coordinates defined for the
        classroom.  We assume that the floor_homography has already been calculated
        before calling this function to map floor pixel coordinates.
        
        Paramters
        ---------
        src_pixels - The [x,y] pixel point we want to map into the cartesian coordinate system
           defined for the classroom.  We assume this is a regular python list that is given, though
           if we need a numpy array instead we should modify this function to expect an array.
           
        Returns
        -------
        [x,y] - Returns the x,y position of the src pixel in the cartesian coordinate system of the
            classroom.
        """
        src_pixels = np.array(src_pixels + [1]) # append 1 for matrix multiplication operations
        dst_coordinates = np.dot(self.floor_homography, src_pixels)
        
        # need to scale by the scale factor
        scale = dst_coordinates[2]
        dst_coordinates = dst_coordinates / scale
        
        # only return the resulting scaled cartesian coordinates of the indicated pixel
        return dst_coordinates[:-1]
        
        
    def addBoard(self, board_obj, position):
        if len(position) == 2:
            self.ROI = (position[0] + board_obj.cx, position[1]+ board_obj.cy)
        else:
            self.ROI = (self.x, self.y)
            
        print('Board added at position {}'.format(self.ROI))
        #self.objects['board_obj'] = (position[0], position[1], self.Point.z)
        return self.ROI

class Board:
    def __init__(self, width, height):
        self.height = height
        self.width = width
        self.depth = 0
        self.cx, self.cy = width//2, height//2
```

```python
# This cell runs on every frame of the stream which runs an 
# object tracker and a face detector at next frame under 
# certain conditions and constrainst as described below.


cam = cv2.VideoCapture(0)
while True:

    _,frame = cam.read()
    roi = frame
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # detect dlib face rectangles
    faces = face_detector(gray, 0)

    for k, face in enumerate(faces):

        eyes = []

        # convert dlib rect to a bounding box
        x,y,w,h = face_utils.rect_to_bb(face)

        # draw rectangle around each face detected
        cv2.rectangle(frame,(x,y),(255,0,0), thickness=1)

        # We c 

        _face = predictor(gray, face)
        _face = face_utils.shape_to_np(_face)
        print(_face)

        # Eye indexes
        leftEye = _face[lEye_l:lEye_r]
        #leftEye_mid = LA.norm(eye[0], eye[3],1)
        #rightEye_mid = LA.norm(eye[0], eye[3],1)
        rightEye = _face[rEye_l:rEye_r]

        Eye1 = Eye(leftEye)
        Eye2 = Eye(rightEye)

        eyes.append(Eye1)  #wrap in a list
        eyes.append(Eye2)


        # loop through both eyes
        for index, eye in enumerate(eyes):

            eye_EAR = Ear(eye)
            single_eye_state = []  # first entry is eye index, second entry is the closed or eye direction state

            left_side_eye = eye[0]  # left edge of eye
            right_side_eye = eye[3]  # right edge of eye
            top_side_eye = eye[1]  # top side of eye
            bottom_side_eye = eye[4]  # bottom side of eye

            # calculate height and width of dlib eye keypoints
            eye_width = right_side_eye[0] - left_side_eye[0]
            eye_height = bottom_side_eye[1] - top_side_eye[1]

            # create bounding box with buffer around keypoints
            eye_x1 = int(left_side_eye[0] - 0 * eye_width)  # .25 works well too
            eye_x2 = int(right_side_eye[0] + 0 * eye_height)  # .75 works well too

            eye_y1 = int(top_side_eye[1] - 1 * eye_height)
            eye_y2 = int(bottom_side_eye[1] + 1 * eye_height)

            # draw bounding box around eye roi
            cv2.rectangle(frame,(eye_x1, eye_y1), (eye_x2, eye_y2),(0,255,0),2)

            # draw the circles for the eye landmarks
            for i in eye:
                cv2.circle(frame, tuple(i), 3, (0, 0, 255), -1)

            # d=10920.0/float(w)

            roi = frame[eye_y1:eye_y2,eye_x1:eye_x2]

            #  ---------    check if eyes open   -------------  #

            # state is open/close, or direction looking
            eye_state = None

            if eye_EAR > 0.25:

                #  ---------    find center of pupil   -------------  #

                gray=cv2.cvtColor(roi,cv2.COLOR_BGR2GRAY)  # grey scale convert
                blur = cv2.medianBlur(gray,5) # blue image to find the iris better
                equ = cv2.equalizeHist(blur)  # ie, improve contrast by spreading the range over the same window of intensity
                thres=cv2.inRange(equ,0,15)  # threshold the contour edges, higher number means more will be black
                kernel = np.ones((3,3),np.uint8)  # placeholder

            #     #/------- removing small noise inside the white image ---------/#
                dilation = cv2.dilate(thres,kernel,iterations = 2)
            #     #/------- decreasing the size of the white region -------------/#
                erosion = cv2.erode(dilation,kernel,iterations = 3)
            #     #/-------- finding the contours -------------------------------/#
                image, contours, hierarchy = cv2.findContours(erosion,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
            #     #--------- checking for 2 contours found or not ----------------#

                pupil_found = None

                if len(contours)==2 :
                    # print('2 contours found')
                    pupil_found = True

                    img = cv2.drawContours(roi, contours, 1, (0,255,0), 3)
                    #------ finding the centroid of the contour ----------------#
                    M = cv2.moments(contours[1])

                    if M['m00']!=0:
                        cx = int(M['m10']/M['m00'])
                        cy = int(M['m01']/M['m00'])
                        cv2.line(roi,(cx,cy),(cx,cy),(0,0,255),3)
                        # print(cx,cy)

                #we

                if len(contours)==1:
                    pupil_found = True
                    # print('only 1 contour found ------- ')

                    img = cv2.drawContours(roi, contours, 0, (0,255,0), 3)

                    #------- finding centroid of the contour ----#
                    M = cv2.moments(contours[0])
                    if M['m00']!=0:
                        cx = int(M['m10']/M['m00'])
                        cy = int(M['m01']/M['m00'])
                        # print(cx,cy)
                        cv2.line(roi,(cx,cy),(cx,cy),(0,0,255),3)

                if pupil_found:
                    # find ratio of distance from each side of the eye bounding box
                    # to get quantify direction of pupil

                    width_ratio = cx / eye_width
                    height_ratio = cy / (eye_y2 - eye_y1)  # make sure to use bounding box height

                # eyes are opened, but pupils not found
                else:
                    print('Pupil not found')
                    single_eye_state.append(index)
                    single_eye_state.append('No pupil found')


        # end loop for one face

    cv2.imshow("frame",frame)
    # if the `q` key was pressed, break from the loop
    key = cv2.waitKey(1) & 0xFF
    if key == ord("q"):
        break
vs.release()
# print("accurracy=",(float(numerator)/float(numerator+denominator))*100)
cv2.destroyAllWindows()
```
