```python
import time
import psutil
import matplotlib.pyplot as plt
help(psutil)
```

```python
%matplotlib notebook
plt.rcParams['animation.html'] = 'jshtml'
fig = plt.figure()
ax = fig.add_subplot()
fig.show()
```

```python
i = 0
x, y = [],[]

while True:
    x.append(i)
    y.append(psutil.cpu_percent())
    
    ax.plot(x,y, color='b')
    fig.canvas.draw()
    ax.set_xlim(left=max(0,i-50),right=i+50)
    time.sleep(0.01)
    i+=1
```
